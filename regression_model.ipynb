```python
import turtle
import random

# تنظیم صفحه
screen = turtle.Screen()
screen.bgcolor("white")
screen.title("شبیه‌سازی رگرسیون خطی")

# لاک‌پشت برای کشیدن نقاط
dot = turtle.Turtle()
dot.penup()
dot.hideturtle()
dot.speed(0)

# لاک‌پشت برای کشیدن خط
line_drawer = turtle.Turtle()
line_drawer.penup()
line_drawer.hideturtle()
line_drawer.pensize(2)
line_drawer.color("red")

# تابع مقیاس دهی به داده‌ها برای نمایش در turtle
def scale(x, y):
    return x - 300, y - 200  # انتقال محور به مرکز صفحه

# تولید نقاط تصادفی با پراکندگی حول خط y = 0.5x + 20
points = []
for _ in range(50):
    x = random.randint(0, 600)
    y = int(0.5 * x + 20 + random.gauss(0, 20))  # نویز با توزیع گاوسی
    points.append((x, y))

# رسم نقاط
for x, y in points:
    sx, sy = scale(x, y)
    dot.goto(sx, sy)
    dot.dot(5, "blue")

# رسم خط رگرسیون فرضی y = 0.5x + 20
x_start, x_end = 0, 600
y_start = int(0.5 * x_start + 20)
y_end = int(0.5 * x_end + 20)

sx1, sy1 = scale(x_start, y_start)
sx2, sy2 = scale(x_end, y_end)

line_drawer.goto(sx1, sy1)
line_drawer.pendown()
line_drawer.goto(sx2, sy2)

# باز نگه داشتن پنجره
screen.mainloop()
```
