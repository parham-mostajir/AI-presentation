```python
import turtle 
from sklearn.datasets import make_blobs
import matplotlib as plt
import seaborn as sb
import sklearn
import numpy as np
```

```python
np.random.seed(0)
```

```python
x,y = make_blobs(n_samples=2000)
print(x[:,0])
```

```python
sb.scatterplot(x=x[:,0],y=x[:,1])
```

```python
def clr(): 
    color = np.random.choice(["#dd674e","#4f4745","#2442ba"])
    return(color)
```

```python
tur = turtle.Turtle()
tur.ht()
tur.speed(0)

screen = turtle.Screen()
screen.title("KNN Data Visualization")
# screen.setup(800, 600)


for input in range(500):
    x_1 = x[input,0]*100
    x_2 = x[input,1]*100 
    tur.penup()
    tur.goto(x_1,x_2)
    tur.pendown()
    tur.fillcolor(clr())
    tur.begin_fill()
    tur.circle(5)
    tur.end_fill()
    
    
tur.penup()
tur.goto(100,-100)
tur.pendown()
tur.fillcolor("#00FF00")
tur.begin_fill()
tur.pencolor("white")
tur.circle(20)
tur.end_fill()
turtle.done()
```
